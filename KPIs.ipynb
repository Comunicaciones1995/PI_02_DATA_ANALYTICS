# INDICADOR CLAVE DE DESEMPEÑO O KPI

PARA ABORDAR ESTA PARTE DEL PROYECTO, PRIMERO SE DEBE CALCULAR Y GRAFICAR LOS DOS (02) KPIs PROPUESTOS RELACIONADOS CON LA SEGURIDAD VIAL DE LA CIUDAD AUTÓNOMA DE BUENOS AIRES (CABA), LUEGO SE DEBERÁ PROPONER, MEDIR Y GRAFICAR TRES (03) KPIs RELEVANTES PARA LA SEGURIDAD VIAL.

```python
# IMPORTACIÓN DE LIBRERIAS...

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# CARGA DE LOS PRINCIPALES DATAFRAMES PARA LOS KPIs...

hechos_homicidios = pd.read_parquet("dataset_limpio/hechos_homicidios.parquet")
victimas_homicidios = pd.read_parquet("dataset_limpio/victimas_homicidios.parquet")
hechos_lesiones = pd.read_parquet("dataset_limpio/hechos_lesiones.parquet")
victimas_lesiones = pd.read_parquet("dataset_limpio/victimas_lesiones.parquet")

print("El proceso se realizó con éxito...")
```

## 1 - PRIMER KPI "Reducir en un 10% la tasa de homicidios en siniestros viales de los últimos seis meses, en CABA, en comparación con la tasa de homicidios en siniestros viales del semestre anterior."

```python
# Filtrar solo los registros del año 2021
df_datos_2021 = hechos_homicidios[hechos_homicidios['AÑO'] == 2021]
```

```python
# Población total
poblacion_total = 3120612

# Filtrar los datos para el primer semestre de 2021
primer_semestre = df_datos_2021[(df_datos_2021['AÑO'] == 2021) & (df_datos_2021['MES'] >= 1) & (df_datos_2021['MES'] <= 6)]

# Filtrar los datos para el segundo semestre de 2021
segundo_semestre = df_datos_2021[(df_datos_2021['AÑO'] == 2021) & (df_datos_2021['MES'] >= 7) & (df_datos_2021['MES'] <= 12)]

# Contar la cantidad de valores en la columna 'ID_hecho' para el primer semestre
cantidad_primer_semestre = primer_semestre['ID'].count()

# Contar la cantidad de valores en la columna 'ID_hecho' para el segundo semestre
cantidad_segundo_semestre = segundo_semestre['ID'].count()

# Calcular la tasa de homicidios en siniestros viales para el primer semestre
tasa_primer_semestre = (cantidad_primer_semestre / poblacion_total) * 100000

# Calcular la tasa de homicidios en siniestros viales para el segundo semestre
tasa_segundo_semestre = (cantidad_segundo_semestre / poblacion_total) * 100000

# Calcular la diferencia entre las tasas del segundo semestre y el primer semestre
diferencia_tasas = tasa_primer_semestre - tasa_segundo_semestre

# Calcular el porcentaje de reducción
porcentaje_reduccion = (diferencia_tasas / tasa_primer_semestre) * 100

# Imprimir los resultados
print("\n Tasa de homicidios en siniestros viales para el primer semestre de 2021:", tasa_primer_semestre)
print("\n Tasa de homicidios en siniestros viales para el segundo semestre de 2021:", tasa_segundo_semestre)
print("\n Diferencia en tasas de homicidios en siniestros viales entre el primer y segundo semestre de 2021:", diferencia_tasas)
print("\n Porcentaje de reducción entre el primer y segundo semestre de 2021:", porcentaje_reduccion, "%")
```

```python
# Datos
semestres = ['Primer Semestre', 'Segundo Semestre']
tasas = [tasa_primer_semestre, tasa_segundo_semestre]

# Crear un gráfico de barras
plt.figure(figsize=(8, 6))
plt.bar(semestres, tasas, color=['blue', 'green'])
plt.title('Tasa de homicidios en siniestros viales por semestre (2021)')
plt.xlabel('Semestre')
plt.ylabel('Tasa de homicidios por 100,000 habitantes')
plt.ylim(0, max(tasas) + 1)
plt.grid(axis='y')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

### CONCLUSIONES FINALES DEL PRIMER KPI...

EN LA PRIMERA KPI SE CALCULÓ LA TASA DE HOMISIDIOS DEL PRIMER Y SEGUNDO SEMESTRE DEL AÑO 2021 EN LA CIUDAD AUTÓNOMA DE BUENOS AIRES Y LUEGO SE CALCULÓ EL PORCENTAJE DE REDUCCIÓN DEL SEGUNDO SEMESTRE EN COMPARACIÓN AL PRIMER SEMESTRE, DANDO UN VALOR DEL 22,2 % DE REDUCCIÓN.

## 2 - SEGUNDO KPI "Reducir en un 7% la cantidad de accidentes mortales de motociclistas en el último año, en CABA, respecto al año anterior."

```python
# Filtrar solo los registros del año 2021
df_datos_MOTOS_2021 = hechos_homicidios[hechos_homicidios['AÑO'] == 2021]
df_datos_MOTOS_2020 = hechos_homicidios[hechos_homicidios['AÑO'] == 2020]
```

```python
# Contar la cantidad de valores "MOTO" en la columna 'VICTIMA' para el año 2020
count_accidentes_2020 = len(df_datos_MOTOS_2020[df_datos_MOTOS_2020['VICTIMA'] == 'MOTO'])

# Contar la cantidad de valores "MOTO" en la columna 'VICTIMA' para el año 2021
count_accidentes_2021 = len(df_datos_MOTOS_2021[df_datos_MOTOS_2021['VICTIMA'] == 'MOTO'])

# Calcular el KPI
if count_accidentes_2020 != 0:
    kpi = ((count_accidentes_2020 - count_accidentes_2021) / count_accidentes_2020) * 100
else:
    kpi = None

# Imprimir la cantidad de valores "MOTO" en la columna 'VICTIMA' para cada año
print(f"Cantidad de accidentes en moto en el año 2020: {count_accidentes_2020}")
print(f"Cantidad de accidentes en moto en el año 2021: {count_accidentes_2021}")
```

```python
# Crear un gráfico de barras para representar el KPI
if kpi is not None:
    plt.figure(figsize=(10, 12))
    plt.bar(['Año 2020', 'Año 2021'], [count_accidentes_2020, count_accidentes_2021], color=['blue', 'green'])
    plt.title('Cantidad de Accidentes Mortales de Motociclistas (2020 - 2021)')
    plt.xlabel('Año')
    plt.ylabel('Número de Accidentes Mortales de Motociclistas')
    plt.ylim(0, max(count_accidentes_2020, count_accidentes_2021) + 1)
    plt.text(0, count_accidentes_2020, f'{count_accidentes_2020} Accidentes', ha='center', va='bottom')
    plt.text(1, count_accidentes_2021, f'{count_accidentes_2021} Accidentes', ha='center', va='bottom')
    plt.grid(axis='y')

    # Mostrar el KPI en el gráfico
    plt.text(1.15, count_accidentes_2020, f'KPI: {kpi:.2f}%', ha='left', va='center', fontsize=12, color='red')

    # Mostrar el gráfico
    plt.tight_layout()
    plt.show()
else:
    print("No hay datos para el año 2020.")
```

### - CONCLUSIONES FINALES DEL SEGUNDO KPI...

EN LA SEGUNDA KPI SE CALCULÓ LA cantidad de accidentes mortales de motociclistas en siniestros viales ENTRE EL AÑO 2021 Y EL AÑO 2020 EN LA CIUDAD AUTÓNOMA DE BUENOS AIRES Y LUEGO SE CALCULÓ EL CAMBIO EN LOS VALORES DE ACCIDENTES MORTALES EN MOTO USANDO LA FÓRMULA CITADA, DANDO UN VALOR DEFINITIVO DEL 70,3 % EN COMPARACIÓN AL 2020, ES DECIR QUE LOS ACCIDENTES POR MOTO AUMENTARON UN 70,3 %. ESTO MUESTRA QUE NO SE PRODUJO UNA REDUCCIÓN EN EL PORCENTAJE DE ACCIDENTES POR MOTOCICLETA SEGÚN LA CONSIGNA.

## 3 - TERCER KPI "Reducir en un 15% la tasa de MORTALIDAD en siniestros viales producidos por AUTOS en los últimos seis meses, en CABA, en comparación con la tasa de MORTALIDAD en siniestros viales de AUTOS del semestre anterior."

```python
# Filtrar los últimos seis meses del año 2020 y los primeros seis meses del año 2021
df_datos_AUTO_ultimos_seis_meses_2020 = hechos_homicidios[(hechos_homicidios['AÑO'] == 2020) & (hechos_homicidios['MES'] >= 7)]
df_datos_AUTO_primeros_seis_meses_2021 = hechos_homicidios[(hechos_homicidios['AÑO'] == 2021) & (hechos_homicidios['MES'] <= 6)]

# Contar la cantidad de valores "AUTO" en la columna 'VICTIMA' para los últimos seis meses del año 2020
count_accidentes_ultimos_seis_meses_2020 = len(df_datos_AUTO_ultimos_seis_meses_2020[df_datos_AUTO_ultimos_seis_meses_2020['VICTIMA'] == 'AUTO'])

# Contar la cantidad de valores "AUTO" en la columna 'VICTIMA' para los primeros seis meses del año 2021
count_accidentes_primeros_seis_meses_2021 = len(df_datos_AUTO_primeros_seis_meses_2021[df_datos_AUTO_primeros_seis_meses_2021['VICTIMA'] == 'AUTO'])

# Calcular el KPI
if count_accidentes_ultimos_seis_meses_2020 != 0:
    kpi = ((count_accidentes_ultimos_seis_meses_2020 - count_accidentes_primeros_seis_meses_2021) / count_accidentes_ultimos_seis_meses_2020) * 100
else:
    kpi = None

# Imprimir la cantidad de valores "AUTO" en la columna 'VICTIMA' para cada período
print(f"Cantidad de accidentes en auto en los últimos seis meses de 2020: {count_accidentes_ultimos_seis_meses_2020}")
print(f"Cantidad de accidentes en auto en los primeros seis meses de 2021: {count_accidentes_primeros_seis_meses_2021}")

# Imprimir el KPI
if kpi is not None:
    print(f"KPI: Reducción del {kpi:.2f}% en la tasa de mortalidad de siniestros viales de autos.")
else:
    print("No hay datos para los últimos seis meses de 2020.")
```

```python
# Crear un gráfico de barras para representar el KPI
if kpi is not None:
    plt.figure(figsize=(10, 12))
    plt.bar(['Últimos 6 meses de 2020', 'Primeros 6 meses de 2021'], [count_accidentes_ultimos_seis_meses_2020, count_accidentes_primeros_seis_meses_2021], color=['blue', 'green'])
    plt.title('Cantidad de Accidentes Mortales de Motociclistas (Últimos 6 meses de 2020 - Primeros 6 meses de 2021)')
    plt.xlabel('Período')
    plt.ylabel('Número de Accidentes Mortales de Motociclistas')
    plt.ylim(0, max(count_accidentes_ultimos_seis_meses_2020, count_accidentes_primeros_seis_meses_2021) + 1)
    plt.text(0, count_accidentes_ultimos_seis_meses_2020, f'{count_accidentes_ultimos_seis_meses_2020} Accidentes', ha='center', va='bottom')
    plt.text(1, count_accidentes_primeros_seis_meses_2021, f'{count_accidentes_primeros_seis_meses_2021} Accidentes', ha='center', va='bottom')
    plt.grid(axis='y')

    # Mostrar el KPI en el gráfico
    plt.text(1.15, count_accidentes_ultimos_seis_meses_2020, f'KPI: {kpi:.2f}%', ha='left', va='center', fontsize=12, color='red')

    # Mostrar el gráfico
    plt.tight_layout()
    plt.show()
else:
    print("No hay datos para los últimos seis meses de 2020.")
```

### - CONCLUSIONES FINALES DEL TERCER KPI...

EN EL TERCER KPI SE CALCULÓ LA cantidad de accidentes mortales de AUTOS en siniestros viales ENTRE EL AÑO 2021 Y EL AÑO 2020 EN LA CIUDAD AUTÓNOMA DE BUENOS AIRES Y LUEGO SE CALCULÓ EL CAMBIO EN LOS VALORES DE ACCIDENTES MORTALES EN AUTO, DANDO UN VALOR DEFINITIVO DEL 37,50 % EN COMPARACIÓN AL 2020, ES DECIR QUE LOS ACCIDENTES POR AUTO TUVO VARIACIONES ENTRE LOS CITADOS AÑOS. ESTO MUESTRA QUE SE PRODUJO UNA REDUCCIÓN EN EL PORCENTAJE DE ACCIDENTES POR AUTOS.

## 4 - CUARTO KPI "Reducir un 8% la tasa de LESIONES en siniestros viales producidos POR EL PERSONAL CON 30 AÑOS DE EDAD entre el año 2020 y el año 2021 en la provincia de Buenos Aires."

```python
# Filtrar los últimos seis meses del año 2020 y los primeros seis meses del año 2021
df_datos_SINIESTROS_2020 = victimas_lesiones[(victimas_lesiones['AÑO'] == 2020) & (victimas_lesiones['EDAD_VICTIMA'] == 30)]
df_datos_SINIESTROS_2021 = victimas_lesiones[(victimas_lesiones['AÑO'] == 2021) & (victimas_lesiones['EDAD_VICTIMA'] == 30)]

# Contar la cantidad de valores "AUTO" en la columna 'VICTIMA' para los últimos seis meses del año 2020
count_datos_SINIESTROS_2020 = len(df_datos_SINIESTROS_2020[df_datos_SINIESTROS_2020['EDAD_VICTIMA'] == 30])

# Contar la cantidad de valores "AUTO" en la columna 'VICTIMA' para los primeros seis meses del año 2021
count_datos_SINIESTROS_2021 = len(df_datos_SINIESTROS_2021[df_datos_SINIESTROS_2021['EDAD_VICTIMA'] == 30])

# Calcular el KPI
if count_datos_SINIESTROS_2020 != 0:
    kpi = ((count_datos_SINIESTROS_2020 - count_datos_SINIESTROS_2021) / count_datos_SINIESTROS_2020) * 100
else:
    kpi = None

# Imprimir la cantidad de valores "AUTO" en la columna 'VICTIMA' para cada período
print(f"Cantidad de accidentes por lesiones del personal de 30 años en el 2020: {count_datos_SINIESTROS_2020}")
print(f"Cantidad de accidentes por lesiones del personal de 30 años en el 2021: {count_datos_SINIESTROS_2021}")

# Imprimir el KPI
if kpi is not None:
    print(f"KPI: Reducción del {kpi:.2f}% en la tasa de accidentes por lesiones del personal de 30 años de edad.")
else:
    print("No hay datos para los años seleccionados.")
```

```python
# Crear un gráfico de barras para representar el KPI
if kpi is not None:
    plt.figure(figsize=(12, 14))  # Ajustar el tamaño del gráfico aquí
    plt.bar(['Año 2020', 'Año 2021'], [count_datos_SINIESTROS_2020, count_datos_SINIESTROS_2021], color=['red', 'yellow'])
    plt.title('Cantidad de Accidentes por lesiones del personal de 30 años de edad (Año 2020 - Año 2021)')
    plt.xlabel('Período')
    plt.ylabel('Número de Accidentes por lesiones del personal de 30 años de edad')
    plt.ylim(0, max(count_datos_SINIESTROS_2020, count_datos_SINIESTROS_2021) * 1.2)  # Ajustar el factor de escala aquí
    plt.text(0, count_datos_SINIESTROS_2020, f'{count_datos_SINIESTROS_2020} Accidentes', ha='center', va='bottom')
    plt.text(1, count_datos_SINIESTROS_2021, f'{count_datos_SINIESTROS_2021} Accidentes', ha='center', va='bottom')
    plt.grid(axis='y')

    # Mostrar el KPI en el gráfico
    plt.text(1.15, count_datos_SINIESTROS_2020, f'KPI: {kpi:.2f}%', ha='left', va='center', fontsize=12, color='red')

    # Mostrar el gráfico
    plt.tight_layout()
    plt.show()
else:
    print("No hay datos para los últimos seis meses de 2020.")
```

### - CONCLUSIONES FINALES DEL CUARTO KPI KPI...

EN EL CUARTO KPI SE CALCULÓ LA cantidad de accidentes por lesiones del personal de 30 años de edad ENTRE EL AÑO 2021 Y EL AÑO 2020 EN LA CIUDAD AUTÓNOMA DE BUENOS AIRES DONDE SE CALCULÓ EL CAMBIO EN LOS VALORES DE ACCIDENTES, DANDO UN VALOR DEFINITIVO DEL -47,59 % EN COMPARACIÓN AL 2020, ES DECIR QUE LOS ACCIDENTES POR LESIONES TUVO VARIACIONES ENTRE LOS CITADOS AÑOS. ESTO MUESTRA QUE NO SE PRODUJO UNA REDUCCIÓN EN EL PORCENTAJE DE ACCIDENTES POR LESIONES AL PERSONAL DE 30 AÑOS DE EDAD.
